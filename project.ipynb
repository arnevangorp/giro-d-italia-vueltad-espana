1. DATA


```python
import requests
from bs4 import BeautifulSoup
import re

# ---- CONFIG ----
URL = "https://www.procyclingstats.com/race/giro-d-italia/2025/stage-21/result/result"
YEAR = 2025

# ---- SCRAPE HTML ----
response = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'})
soup = BeautifulSoup(response.text, "html.parser")

# ---- PARSE RESULT TABLE ----
table = soup.find("table", class_="results")
rows = table.find_all("tr")  # alle rijen

data = []

for row in rows:
    cols = row.find_all("td")
    if len(cols) < 4:
        continue  # negeer lege of niet-relevante rijen

    # Rank
    rank = cols[0].get_text(strip=True)

    # Rider
    rider_cell = row.find("td", class_="ridername")
    rider_link = rider_cell.find("a") if rider_cell else None
    if rider_link:
        name_parts = rider_link.stripped_strings
        rider = " ".join(name_parts)
    else:
        rider = rider_cell.get_text(strip=True) if rider_cell else ""

    # Team
    team_cell = row.find("td", class_="cu600")
    team = team_cell.get_text(strip=True) if team_cell else ""

    # Tijd / Gap
    time_cell = row.find("td", class_="time")
    gap_cell = row.find("td", class_="gap")

    if time_cell and time_cell.get_text(strip=True):
        # winnaar heeft meestal geen gap, de tijd staat in 'time'
        time_clean = re.sub(r"[^0-9:]", "", time_cell.get_text(strip=True))
        gap_clean = "-" if not gap_cell else re.sub(r"[^0-9:+']", "", gap_cell.get_text(strip=True))
    else:
        # anderen hebben 'gap' en soms lege tijd
        time_clean = ""
        gap_clean = re.sub(r"[^0-9:+']", "", gap_cell.get_text(strip=True)) if gap_cell else "-"

    data.append({
        "Rank": rank,
        "Rider": rider,
        "Team": team,
        "Year": YEAR,
        "Time": time_clean,
        "Gap": gap_clean
    })

# ---- TERMINAL OUTPUT ----
header = f"{'Rank':<5} {'Rider':<25} {'Team':<30} {'Year':<6} {'Time':<10} {'Gap':<10}"
print(header)
print("-" * len(header))

for row in data:
    print(f"{row['Rank']:<5} {row['Rider']:<25} {row['Team']:<30} {row['Year']:<6} {row['Time']:<10} {row['Gap']:<10}")
```
